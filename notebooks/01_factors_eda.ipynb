# Factor EDA

Explore the monthly Fama-French factors used by the project pipeline.

```python
from pathlib import Path
import sys

PROJECT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(PROJECT))

import pandas as pd
from IPython.display import Image, display
from src.config import ProjectConfig
from src.data import fetch_fama_french_factors

config = ProjectConfig()
factors = fetch_fama_french_factors(config.start, config.end, config.data_dir)
display(factors.tail())
```

```python
annualized = pd.DataFrame({
    'Mean Ann.': factors[['MKT_RF', 'SMB', 'HML', 'RMW', 'CMA', 'MOM']].mean() * 12,
    'Vol Ann.': factors[['MKT_RF', 'SMB', 'HML', 'RMW', 'CMA', 'MOM']].std() * (12 ** 0.5),
})
annualized['Sharpe'] = annualized['Mean Ann.'] / annualized['Vol Ann.']
display(annualized)
display(factors[['MKT_RF', 'SMB', 'HML', 'RMW', 'CMA', 'MOM']].corr())
```
